### Variables in Both Rows AND Columns

1.What happens when we have variables in both rows and columns.


2.How to know that there is data everywhere?


3.If you end up with the vast majority of the row is duplicated, variables are probably being stored in both rows and columns.


4.So variables are being stored in a row.

```python
weather = pd.read_csv('https://raw.githubusercontent.com/chendaniely/pydatadc_2018-tidy/master/data/weather.csv')
```

```python
#Getting year, month, day, and temperature max/min
weather
```

How do we pivot the values back up? tmax and tmin are column names and headers



```python
#Weather_melt OR weather.melt
weather_melt = pd.melt(
    weather,
    id_vars = ['id','year','month','element'],
    var_name='day',
    value_name='temp'
)
```

```python
#Melt the weather_df
weather_melt.head()
```

Got the data fixed, but still want it's a min or max temperature. If we wanted to do a model for year, month, and day, could we do it? We want the min or max temperature, so that we could do a model, but we have no way to specify this currently.
With pivot, we can specify the min, mean, max, etc.


```python
#Melting and pivoting the table
weather_tidy = weather_melt.pivot_table(
    index = ['id', 'year', 'month', 'day'],
    columns = 'element', #What to pivot on
    values = 'temp', #Where the values go
)
```

```python
weather_tidy.head()
```

```python
#If we want a regular, flat dataframe, we can use reset_index
weather_tidy.reset_index()
```
