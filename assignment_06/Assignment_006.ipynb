#### Write a function named cleanse_year() that takes a string as input and removes any asterisks in the string and return the cleansed string. Test the function by using test strings (for example, "1989*", "20*10") 

```python

def cleanse_year(old_year):
    year = old_year.replace("*","")
    return year
```

Test the function by using test strings:

```python
cleanse_year("1989*")
```

```python
cleanse_year("20*10")
```

#### Cleanse the column "Census Year" using the function cleanse_year() and Pandas's apply() function.


```python
df1['Census Year'] = df1['Census Year'].apply(cleanse_year)
df1
```

#### Repeat this process for column "Total Population", "Census Cost", "Average Cost Per Person". (define, test, and apply the function)



```python

import math
# defining
def cleanse_cost(raw_cost):
    cost = raw_cost.lstrip("$").replace(",","")
    if  cost.find("Billion") != -1:          # if it conatins "Billion":
        cost = cost.replace('Billion','')
        cost = float(cost)* math.pow(10,8)
    elif raw_cost.find("Million") != -1:     # if it contains "Million": 
        cost = cost.replace('Million','')
        cost = float(cost)* math.pow(10,6)
    return str(cost)

#Testing
cleanse_cost("13 Billion")
```

Applying cleansed Census Cost to the dataframe



```python
df1["Census Cost"] = df1["Census Cost"].apply(cleanse_cost)
df1
```

Cleansing Total Population



```python
#defining 
def cleanse_population(population):
  return int(population.replace(',',''))

#testing

cleanse_population("123,000")
```

Applying cleansed Total Population to the dataframe



```python
df1['Total Population'] = df1['Total Population'].apply(cleanse_population)
df1
```

Cleansing Average Cost Per Person



```python
#defining 

def cleanse_avgcost(cost):
  cost = cost.replace('$','')
  cost = cost.replace('\t','')
  cost = cost.replace('\n','')
  if 'cents' in cost:
    # print("Cost ",cost)
    cost = cost.replace('cents','')
    cost = float(cost)/100
  return cost

# testing
cleanse_avgcost("1.6 cents")
```

Applying Average Cost Per Person to the dataframe



```python
df1['Average Cost Per Person'] = df1['Average Cost Per Person'].apply(cleanse_avgcost)
df1
```
