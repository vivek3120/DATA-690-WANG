```python
x = list()

i = 0

while(10>i):

    try:

        num = int(input(""))        

    except :

        print('Please enter numericals only')

    else:

        x.append(num)

        i += 1


 

def minimum(x):

    minvalue = x[0]

    for i in x:

        if i < minvalue:

            minvalue = i

    return minvalue

 

def maximum(x):

    maxvalue = x[0]

    for i in x:

        if i > maxvalue:

            maxvalue = i

    return maxvalue

 

def Mean(x):

    totalsum = 0

    length = 0

    for i in x:

        totalsum = totalsum + i

        length = length + 1

    return (totalsum/length)

 
def Range(x):

    rangelist = list()

    rangelist.append(minimum(x))

    rangelist.append(maximum(x))

    return rangelist


def Variance(x):
    total_Sum=0
    count=0
    for num in x:
        res=num-Mean(x)
        total_Sum+=res*res
        count+=1
    Variance_Of_Numbers=total_Sum/(count-1)  
    return Variance_Of_Numbers

def Standard_Deviation(x):
    standard_Deviation= Variance(x) ** 0.5
    return standard_Deviation
   


print("Numbers which you have entered are:",x)
print("minimum equals to",minimum(x))
print("maximum equals to",maximum(x)) 
print("Range equals to",Range(x))
print("mean equals to",Mean(x))
print("variance equals to",Variance(x))
print("standard deviation equals to",Standard_Deviation(x))
```
