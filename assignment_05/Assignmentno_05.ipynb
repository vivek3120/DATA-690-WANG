Definitions or description of the variables which are used in the new dataframe are:-

UNITID - UNITID is for the id of an institution. And the datatype is integer.


INSTNM - INSTNM variable name  is used for the institute name. And the datatype is autocomplete which means it would be suggesting the names based on previous inputs. 


CITY - CITY variable name is used for place or location where the school is located. And the datatype is autocomplete which means it would be suggesting the names based on previous inputs. 

STABBR - STABBR variable name is used for State postcode where the school is located. And the datatype is string.

ZIP - ZIP variable name is used for the zip code(Zone Improvement Plan) or a postal code. And the datatype is string.

ADM_RATE - ADM_RATE variable name is used for the overall admission rate. And the datatype is float.

UGDS -UGDS variable name is used for enrollment of undergraduate certificate/degree-seeking students. And the datatype is Integer.


TUITIONFEE_IN - TUITIONFEE_IN variable name is used for in-state tuition and fees. And the datatype is Integer.


MN_EARN_WNE_P10 - MN_EARN_WNE_P10 variable name is used for mean earnings of students working and not enrolled 10 years after entry. And the datatype is Integer.
